# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Models Training (Multiple Models) for more models to our machine learning models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings("ignore")

# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Support Vector Classifier": SVC(),
    "Random Forest Classifier": RandomForestClassifier(),
    "K Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Gaussian Naive Bayes": GaussianNB(),
    "AdaBoost Classifier": AdaBoostClassifier(),
    "Gradient Boosting Classifier": GradientBoostingClassifier(),
    "XGBoost Classifier": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
}

# Train and evaluate each model
for name, model in models.items():
    print("="*50)
    print("Model:", name)
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # Predict on test set
    y_pred = model.predict(X_test_scaled)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    # Print metrics
    print("Accuracy:", accuracy)
    print("Classification Report:\n", classification_rep)
    print("Confusion Matrix:\n", conf_matrix)
```

# Recommendation System

```python
import pickle
import numpy as np

# Load the scaler, model, and class names
scaler = pickle.load(open("D://Final Projcts//System//STUDIES-CAREER-RECOMMENDATION-SYSTEM-using-machine-learning//STUDIES-CAREER-RECOMMENDATION-SYSTEM-using-machine-learning/Update models/scaler44.pkl", 'rb'))
model = pickle.load(open("D://Final Projcts//System//STUDIES-CAREER-RECOMMENDATION-SYSTEM-using-machine-learning//STUDIES-CAREER-RECOMMENDATION-SYSTEM-using-machine-learning/Update models/model44.pkl", 'rb'))
class_names = ['Lawyer', 'Doctor', 'Government Officer', 'Artist', 'Social Network Studies',
               'Software Engineer', 'Teacher', 'Business Owner', 'Scientist',
               'Banker', 'Writer', 'Accountant', 'Designer',
               'Construction Engineer', 'Game Developer', 'Stock Investor',
               'Real Estate Developer']

# Define the subject combination names based on the class names
subject_combinations = {
    'history_english': ['Lawyer', 'Government Officer', 'Writer', 'Teacher', 'Artist', 'Social Network Studies'],
    'biology_chemistry': ['Doctor', 'Scientist'],
    'history_geography': ['Social Network Studies'],
    'math_physics': ['Software Engineer', 'Scientist', 'Construction Engineer', 'Game Developer'],
    'math_english': ['Business Owner'],
    'math': ['Banker', 'Accountant', 'Stock Investor'],
    'english': ['Writer', 'Designer', 'Social Network Studies'],
    'biology_chemistry_physics': ['Doctor'],
    'math_geography': ['Real Estate Developer', 'Social Network Studies']
}

# Define careers influenced by high self-study hours based on the class names
high_self_study_careers = [
    'Doctor',
    'Scientist',
    'Software Engineer',
    'Lawyer',
    'Accountant',
    'Teacher',
    'Construction Engineer',
    'Game Developer',
    'Stock Investor'
]

# Reasons for high self-study careers
high_self_study_reasons = [
    "Requires extensive study of medical texts, preparation for exams, and continuous learning of new medical advancements.",
    "Involves in-depth research, analysis of complex data, and staying updated with scientific literature and discoveries.",
    "Demands learning new programming languages, algorithms, and keeping up with rapid technological changes.",
    "Requires thorough understanding of laws, precedents, case studies, and continuous study to stay current with legal changes.",
    "Necessitates detailed study of financial regulations, tax laws, and continuous professional development to maintain certification.",
    "Involves preparation of lesson plans, understanding educational theories, and staying updated with teaching methodologies.",
    "Requires studying complex engineering principles, software development practices, and staying updated with industry standards.",
    "Requires studying complex engineering principles, software development practices, and staying updated with industry standards.",
    "Involves in-depth study of market trends, financial reports, and economic indicators to make informed investment decisions."
]

# Define the key factors affecting career aspirations based on the class names
key_factors = {
    'Math Score': ['Software Engineer', 'Scientist', 'Accountant', 'Construction Engineer'],
    'Biology Score': ['Doctor', 'Scientist'],
    'History Score': ['Government Officer', 'Teacher', 'Writer'],
    'Physics and Chemistry Scores': ['Scientist', 'Doctor', 'Construction Engineer'],
    'English Score': ['Lawyer', 'Writer', 'Teacher', 'Artist', 'Designer', 'Social Network Studies'],
    'Geography Score': ['Real Estate Developer', 'Social Network Studies']
}

def Recommendations(gender, part_time_job, absence_days, extracurricular_activities,
                    weekly_self_study_hours, math_score, history_score, physics_score,
                    chemistry_score, biology_score, english_score, geography_score,
                    total_score, average_score):
    
    # Encode categorical variables
    gender_encoded = 1 if gender.lower() == 'female' else 0
    part_time_job_encoded = 1 if part_time_job else 0
    extracurricular_activities_encoded = 1 if extracurricular_activities else 0

    # Create the original feature array for scaling (14 features)
    feature_array = np.array([[gender_encoded, part_time_job_encoded, absence_days, extracurricular_activities_encoded,
                               weekly_self_study_hours, math_score, history_score, physics_score,
                               chemistry_score, biology_score, english_score, geography_score,
                               total_score, average_score]])
    
    # Scale features
    scaled_features = scaler.transform(feature_array)
    
    recommendations = {}

    # Predict for each subject combination
    for combination in subject_combinations.keys():
        # Calculate the average score for the current combination
        if combination == 'history_english':
            subject_avg_score = (history_score + english_score) / 2
        elif combination == 'biology_chemistry':
            subject_avg_score = (biology_score + chemistry_score) / 2
        elif combination == 'history_geography':
            subject_avg_score = (history_score + geography_score) / 2
        elif combination == 'math_physics':
            subject_avg_score = (math_score + physics_score) / 2
        elif combination == 'math_english':
            subject_avg_score = (math_score + english_score) / 2
        elif combination == 'math':
            subject_avg_score = math_score
        elif combination == 'english':
            subject_avg_score = english_score
        elif combination == 'biology_chemistry_physics':
            subject_avg_score = (biology_score + chemistry_score + physics_score) / 3
        elif combination == 'math_geography':
            subject_avg_score = (math_score + geography_score) / 2
        
        # Predict using the model
        probabilities = model.predict_proba(scaled_features)
        
        # Get top five predicted classes along with their probabilities
        top_classes_idx = np.argsort(-probabilities[0])[:5]
        top_classes_names_probs = [(class_names[idx], probabilities[0][idx]) for idx in top_classes_idx]
        
        # Map the predictions to the relevant education and job fields for each subject combination
        recommendations[combination] = {
            'Subjects': subject_combinations[combination],
            'Recommendations': top_classes_names_probs,
            'Subject Average Score': subject_avg_score
        }

    # Add information about high self-study hours influenced careers
    recommendations['High Self-Study Careers'] = {
        'Careers': high_self_study_careers,
        'Reasons': high_self_study_reasons
    }

    # Add key factors affecting career aspirations
    recommendations['Key Factors'] = key_factors
    
    return recommendations

# Example usage
example_recommendations = Recommendations(
    gender='female',
    part_time_job=True,
    absence_days=5,
    extracurricular_activities=True,
    weekly_self_study_hours=15,
    math_score=90,
    history_score=85,
    physics_score=88,
    chemistry_score=92,
    biology_score=87,
    english_score=91,
    geography_score=84,
    total_score=1200,
    average_score=85
)

# Print recommendations in the specified format
for combination, result in example_recommendations.items():
    if combination in subject_combinations:
        print(f"Subject Combination: {combination}")
        print("Relevant Fields:", ", ".join(result['Subjects']))
        print("Top Recommendations:")
        for job, prob in result['Recommendations']:
            print(f" - {job}: {prob:.2f}")
        print()
        
# Print high self-study careers
if 'High Self-Study Careers' in example_recommendations:
    print("High Self-Study Careers:")
    for career in example_recommendations['High Self-Study Careers']['Careers']:
        print(f" - {career}")
    print("Reasons:")
    for reason in example_recommendations['High Self-Study Careers']['Reasons']:
        print(f" - {reason}")
    print()

# Print key factors affecting career aspirations
if 'Key Factors' in example_recommendations:
    print("Key Factors Affecting Career Aspirations:")
    for factor, careers in example_recommendations['Key Factors'].items():
        print(f"{factor}: {', '.join(careers)}")
    print()
```
